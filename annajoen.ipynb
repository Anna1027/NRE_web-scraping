```python
# import libraries 
from bs4 import BeautifulSoup 
import urllib.request
import csv
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#specify the url 
urlpage = 'https://www.law.umich.edu/special/exoneration/Pages/detaillist.aspx'
```

```python
#query the website and return the html to the variable 'page'
page = urllib.request.urlopen(urlpage)
#parge the html using beautiful soup and store in varible 'soup'
soup = BeautifulSoup(page, 'html.parser')
```
